### Cara Object Oriented Programming (OOP)

```python
x = np.arange(1,20,0.5)
y = x**2
```

```python
x
```

```python
y
```

```python
# Versi OOP
# Versi 1
```

```python
fig = plt.figure()
axis = fig.add_axes([0.1,0.1,0.8,0.8])
axis.plot(x,y)
axis.set_xlabel('Sumbu X')
axis.set_ylabel('Sumbu Y')
axis.set_title('Contoh Grafik')
plt.show()
```

```python
fig = plt.figure()
axis1 = fig.add_axes([0.1,0.1,0.8,0.8])
axis2 = fig.add_axes([0.2,0.5,0.3,0.3])

axis1.plot(x,y, color= 'green')
axis1.set_xlabel('Sumbu X')
axis1.set_ylabel('Sumbu Y')
axis1.set_title('grafik1')
axis2.plot(y,x)
```
