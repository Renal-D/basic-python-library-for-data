```python
x = [10,20,30,40,50]
y = np.array(x)**2
```

```python
plt.plot(x,y)
plt.xlabel('angka')
plt.ylabel('angka pangkat 2')
plt.title('Grafik coba coba')
```

```python
plt.subplot(121)
plt.plot(x,y, color= 'red')
plt.xlabel('Sumbu X1')
plt.ylabel('Sumbu Y1')

plt.subplot(122)
plt.plot(y,x, color= 'green')
plt.xlabel('Sumbu X2')
plt.ylabel('Sumbu Y2')
plt.tight_layout()
plt.show()
```
