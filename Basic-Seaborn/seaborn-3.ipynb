```python
import pandas as pd 
import seaborn as sns
sns.set()
```

```python
dataku = pd.read_csv('kapal_titanic.csv')
```

```python
dataku.head()
```

```python
numeric_data = dataku.select_dtypes(include=['number'])
numeric_data.corr()
```

```python
# Heatmap (matrix plot)
sns.heatmap(numeric_data.corr());
```

```python
sns.heatmap(numeric_data.corr(),annot=True);
```

```python
sns.heatmap(numeric_data.corr(),annot=True,cmap='Greens');
```

```python
terbang = sns.load_dataset('flights')
```

```python
terbang
```

```python
terbang_pivot = terbang.pivot_table(index='month', columns='year', values='passengers')
```

```python
sns.heatmap(terbang_pivot, cmap='Blues');
```

```python
sns.heatmap(terbang_pivot, cmap='Blues', linecolor='White', linewidths='2');
```

```python
sns.clustermap(terbang_pivot); # hasil dari proses clustering 
```

```python
sns.clustermap(terbang_pivot, cmap='coolwarm');
```

```python
sns.clustermap(terbang_pivot, cmap= 'coolwarm', standard_scale=1);
```
