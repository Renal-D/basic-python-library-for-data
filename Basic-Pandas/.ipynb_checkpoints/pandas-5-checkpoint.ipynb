```python
import numpy as np 
```

```python
np.random.seed(100)
```

```python
angka = []
for i in range(0,100):
    angka.append(i)
```

```python
baru = pd.DataFrame(np.random.rand(100,5),angka, ['a','b','c','d','e'])
```

```python
baru.head()
```

```python
baru.plot.area(alpha= 0.5)
```

```python
baru2 = pd.DataFrame(np.random.rand(6,5),[1,2,3,4,5,6], ['a','b','c','d','e'])
```

```python
baru2
```

```python
baru2.plot.bar()
```

```python
baru2.plot.bar(stacked= True)
```

```python
baru2['c'].plot.bar()
```

```python
baru.plot.scatter(x= 'a', y='b', c='e') #atau bisa pake cmap='ini adalah parameter yang ada di matplotlib seperti red, ocean'
```

```python
baru.plot.scatter(x= 'a', y='b', s=baru.d*100, c='a')
```

```python
baru.plot.box()
```

```python
baru.plot.line(y= 'b', cmap= 'ocean', lw= 2, figsize= (10,5))
```

```python
baru.plot.hexbin(x= 'a',y='b', gridsize=10)
```
