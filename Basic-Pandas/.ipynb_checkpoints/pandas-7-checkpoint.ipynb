```python
import pandas as pd
```

```python
satu = pd.DataFrame({'key':['k0','k1','k2','k3','k4'], 
                     'one':[1,2,3,4,5],
                     'two':[6,7,8,9,10]})
```

```python
dua = pd.DataFrame({'key':['k0','k1','k2','k3','k4'], 
                     'three':[1,2,3,4,5],
                     'four':[6,7,8,9,10]})
```

```python
satu
```

```python
dua
```

#### Cara menggabungkan 2 tabel ala ala SQL
menggunakan merge

```python
baru = pd.merge(satu,dua)
```

```python
baru
```

```python
baru2 = pd.merge(satu, dua, how='inner', on='key')
```

```python
baru2
```

```python
baru3 = pd.merge(satu, dua, how='outer', on='key')
```

```python
baru3
```

```python
satu1 = pd.DataFrame({'key1':['K0','K1','K0','K1','K1'], 
                      'key2':['K1','K2','K0','K1','K1'],
                      'A' :['A0','A1','A2','A3','A4'],
                      'B' :['B0','B1','B2','B3','B4']})
```

```python
dua2 = pd.DataFrame({'key1':['K0','K0','K1','K0','K2'], 
                      'key2':['K2','K1','K1','K0','K3'],
                      'C' :['C0','C1','C2','C3','C4'],
                      'D' :['D0','D1','D2','D3','D4']})
```

```python
satu1
```

```python
dua2
```

```python
baru4 = pd.merge(satu1,dua2, how='right', on=['key1','key2'])
```

```python
baru4
```

```python
baru5 = pd.merge(satu1,dua2, how='outer', on=['key1','key2'])
```

```python
baru5
```

```python
hero1 = pd.DataFrame({'Mage':['Silvanna','Keenan','Kagura','Serra','Sona'],
                      'MM':['Serra','Sona','Harith','Ahri','Joy']},
                      index=['K0','K2','K3','K4','K5'])
```

```python
hero2 = pd.DataFrame({'Baron':['Tigreal','Garen','Gwen','Akali','Terizla'],
                      'Supp':['Ruin','Reyna','Mino','Mathilda','Siela']},
                      index=['K0','K1','K2','K3','K5'])
```

```python
hero1
```

```python
hero2
```

```python
hero1.join(hero2) #left join
```

```python
hero1.join(hero2, how='outer') #outer join
```

```python
hero1.join(hero2, how='inner') #inner join
```
